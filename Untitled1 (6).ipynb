```python
Q1: What are the Probability Mass Function (PMF) and Probability Density Function (PDF)? Explain with
an example.
Answer--The Probability Mass Function (PMF) and Probability Density Function (PDF) are
mathematical functions used in probability theory and statistics to describe the distribution 
of random variables, but they are applied to different types of variables: discrete and continuous, respectively.

Probability Mass Function (PMF):

The PMF is a function that provides the probability that a discrete random variable takes on a specific value.
Probability Density Function (PDF):

The PDF is a function that describes the probability distribution of a continuous random variable.
Unlike the PMF, the PDF doesn't directly give the probability of a specific value but gives 
the relative likelihood that the variable falls within a particular range of values.

Q2: What is Cumulative Density Function (CDF)? Explain with an example. Why CDF is used?
Answer--The Cumulative Density Function (CDF) is a function associated with a probability distribution that describes the probability that a random variable 
�
X takes on a value less than or equal to a given value 
�
x. In other words, the CDF gives the cumulative probability up to a certain point.
Q3: What are some examples of situations where the normal distribution might be used as a model?
Explain how the parameters of the normal distribution relate to the shape of the distribution.
Answer--Why CDF is used?
The CDF is used for several reasons:

It provides a complete overview of the distribution of a random variable, including its range, central tendency, and variability.
It allows for easy calculation of probabilities for a range of values, including probabilities of intervals or specific outcomes.
It is used to compare distributions and assess the likelihood of different events occurring.
It is a fundamental tool in statistical inference, hypothesis testing, and modeling.

Q4: Explain the importance of Normal Distribution. Give a few real-life examples of Normal
Distribution.
Answer--The normal distribution, also known as the Gaussian distribution, is one of the most important
probability distributions in statistics and probability theory. Its significance arises from several 
key properties that make it widely applicable in various fields. Here are some reasons why the normal
distribution is important:

Central Limit Theorem: The normal distribution emerges naturally as the limiting distribution of the
sum (or average) of a large number of independent and identically distributed random variables, regardless 
of the underlying distribution. This property is central to inferential statistics and hypothesis testing.

Symmetry and Shape: The normal distribution is symmetric about its mean, with the characteristic bell-shaped curve. 
This symmetry simplifies calculations and allows for easy interpretation of statistical measures such as mean, median, and mode.

Well-Studied Properties: The normal distribution is well-studied, and many statistical techniques and formulas are
derived based on its properties. It facilitates the development of analytical methods for hypothesis testing, 
confidence intervals, and parameter estimation.

Universality: The normal distribution is observed in many natural and human-made phenomena. While some variables
may not exactly follow a normal distribution, many processes tend to exhibit behaviors that approximate it due to the Central Limit Theorem.

Real-life examples of phenomena that follow or approximate a normal distribution include:

Height of Individuals: In a large population, the distribution of heights tends to follow a normal 
distribution, with most individuals clustered around the average height.
Scores on Standardized Tests: Scores on standardized tests like the SAT or IQ tests often approximate
a normal distribution. The majority of test-takers tend to score near the average, with fewer individuals
scoring significantly higher or lower.
Measurement Errors: Errors in measurement devices, such as weighing scales or thermometers, often follow a
normal distribution. The errors may be symmetrically distributed around the true value.
Stock Market Returns: Daily returns of stock prices over a long period often exhibit a distribution that is
approximately normal. This assumption is fundamental in financial modeling and risk analysis.
Biological Variables: Many biological variables, such as blood pressure, cholesterol levels, and reaction 
times, tend to be normally distributed in large populations.

Q5: What is Bernaulli Distribution? Give an Example. What is the difference between Bernoulli
Distribution and Binomial Distribution?
Answer--The Bernoulli distribution is a discrete probability distribution that represents the outcome of
a random experiment or trial with only two possible outcomes: success or failure. It is named after the Swiss mathematician Jacob Bernoulli.

The key characteristics of the Bernoulli distribution are:

It has two possible outcomes: success (usually denoted as 1) and failure (usually denoted as 0).

Q7: Explain uniform Distribution with an example.
Answer--The uniform distribution is a continuous probability distribution that describes a random variable 
where every value within a certain range is equally likely to occur. In other words, all possible outcomes 
have the same probability density.

Key characteristics of the uniform distribution include:

Constant Probability Density: The probability density function (PDF) is constant within the range of
possible values and zero outside that range.
Equal Probability of Outcomes: All values within the range have an equal chance of occurring.
Defined Range: The distribution is defined over a specific interval or range.
Mathematically, the probability density function (PDF) of a uniform distribution is defined as 
Q8: What is the z score? State the importance of the z score.
Answer--The z-score, also known as the standard score, is a statistical measure that indicates how many
standard deviations a data point is from the mean of the dataset. It is a dimensionless quantity
that allows for the comparison of data points from different normal distributions.

The formula to calculate the z-score of a data point 
�
x in a dataset with mean 
�
μ and standard deviation 
�
σ is:
    Where:

�
x is the value of the data point,
�
μ is the mean of the dataset,
�
σ is the standard deviation of the dataset,
�
z is the z-score.
Key points about the z-score and its importance:

Standardization: The z-score standardizes the data, allowing for the comparison of values from different distributions.
By calculating the z-score, data points can be compared to the mean and evaluated in terms of how many standard deviations they are away from the mean.

Interpretation: A z-score of 0 indicates that the data point is exactly at the mean of the distribution.

A positive z-score indicates that the data point is above the mean, while a negative z-score 
indicates that the data point is below the mean.

Outlier Detection: Z-scores can help identify outliers in a dataset. Data points with z-scores that 
are significantly greater than or less than 0 (typically above 2 or below -2) are considered outliers
and may warrant further investigation.

Probability Calculation: Z-scores are often used to calculate probabilities associated with specific
values in a normal distribution. For example, the z-score can be used to determine the probability\
of a data point falling within a certain range or above/below a certain threshold.

Normalization: Z-scores are commonly used in normalization techniques, such as standardizing variables
in machine learning algorithms, where the data is transformed to have a mean of 0 and a standard deviation of 1.
```
